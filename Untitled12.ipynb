```python
def get_book_descriptions(df, isbn_column, client_id, client_secret):
    """
    네이버 책 API를 통해 ISBN 번호를 바탕으로 책 설명을 가져오는 함수.

    Parameters:
    - df (pd.DataFrame): ISBN 번호가 있는 데이터프레임
    - isbn_column (str): ISBN 번호가 포함된 열의 이름
    - client_id (str): 네이버 API 클라이언트 ID
    - client_secret (str): 네이버 API 클라이언트 시크릿

    Returns:
    - pd.DataFrame: description 열이 추가된 데이터프레임
    """
    descriptions = []

    for isbn in df[isbn_column]:
        query = urllib.parse.quote(isbn)

        # 네이버 API URL
        url = f"https://openapi.naver.com/v1/search/book_adv?d_isbn={query}&display=1&start=1&sort=sim"

        # 요청 헤더 설정
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)

        try:
            # API 요청 보내기
            response = urllib.request.urlopen(request)
            rescode = response.getcode()

            if rescode == 200:
                # 응답 본문 읽기
                response_body = response.read()
                items = json.loads(response_body.decode('utf-8')).get("items", [])

                # HTML 태그 제거용 정규 표현식
                remove_tag = re.compile('<.*?>')

                # 책의 description 가져오기 (첫 번째 검색 결과)
                description = re.sub(remove_tag, '', items[0]['description']) if items else "No description available"
                descriptions.append(description)
            else:
                print(f"Error Code: {rescode}")
                descriptions.append("Error in fetching description")

        except urllib.error.HTTPError as e:
            print(f"HTTP Error: {e.code}")
            descriptions.append("HTTP Error")
        except urllib.error.URLError as e:
            print(f"URL Error: {e.reason}")
            descriptions.append("URL Error")
        except Exception as e:
            print(f"General Error: {str(e)}")
            descriptions.append("General Error")

    # DataFrame에 descriptions 열 추가
    df['description'] = descriptions

    return df
```

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def adjust_labels(label_counts, column, threshold):
    """
    Args:
    - label_counts (dict): 각 라벨 내 instance 개수를 담은 딕셔너리
    - column (list or pd.Series): 바꾸고 싶은 DataFrame의 열 (list나 Series 형태 가능)
    - threshold (int): 라벨 내 instance 개수의 임계값

    Returns:
    - pd.Series: 수정된 열
    """
    # 리스트일 경우 Pandas Series로 변환
    if isinstance(column, list):
        column = pd.Series(column)

    def modify_value(label):
        # 값이 문자열이 아닌 경우 문자열로 변환
        if not isinstance(label, str):
            label = str(label)

        # 조건 1: 라벨 내 instance 개수가 threshold를 넘지 못함
        # 조건 2: 라벨이 '.'으로 끝나지 않음
        if label_counts.get(label, 0) < threshold and not label.endswith('.'):
            return label[:-1]  # 맨 마지막 character 삭제
        return label

    # 열(column)에 대해 조건을 적용
    return column.apply(modify_value)

def clean_kdc_nm_column(column):
    """
    KDC_NM 컬럼의 값을 정리합니다. 마지막에 '.'이 있다면 삭제하고, 공백이 포함되어 있다면 제거합니다.

    Args:
        column (list): 데이터프레임의 KDC_NM 컬럼 값 리스트.

    Returns:
        list: 정리된 KDC_NM 값 리스트.
    """
    cleaned_column = []

    for value in column:
        if isinstance(value, str):
            # 공백 제거
            value = value.replace(" ", "")
            # 마지막 문자가 '.'이면 제거
            if value.endswith("."):
                value = value[:-1]
        cleaned_column.append(value)

    return cleaned_column

import pandas as pd

# 변환 불가 값 확인 및 삭제
def drop_invalid(df, column_name):
    """
    Args:
    - df (pd.DataFrame): 데이터프레임
    - column_name (str): 검사할 열 이름

    Returns:
    - pd.DataFrame: 변환 불가능한 값이 제거된 데이터프레임
    """
    def is_valid(value):
        try:
            float(value)  # float 변환 시도
            return True
        except ValueError:
            return False

    # 변환 가능한 값만 필터링
    valid_df = df[df[column_name].apply(is_valid)].reset_index(drop=True)
    return valid_df

def convert_column_decimals(df, column_name, decimals=2):
    """
    Args:
    - df (pd.DataFrame): 데이터프레임
    - column_name (str): 소수점 자리수를 변환할 컬럼 이름
    - decimals (int): 변환할 소수점 자리수 (기본값: 2)

    Returns:
    - pd.DataFrame: 변환된 데이터프레임
    """
    if column_name in df.columns:
        format_string = f"{{:.{decimals}f}}"  # 소수점 자리수 설정
        df[column_name] = df[column_name].apply(
            lambda x: format_string.format(float(x)) if pd.notnull(x) else x
        )
    else:
        raise ValueError(f"Column '{column_name}' not found in the DataFrame.")

    return df
```

```python
import importlib
import numpy as np

import torch
from torch import nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
from transformers import AutoModel, AutoTokenizer
from sklearn.model_selection import train_test_split

# 데이터셋 정의
class BookDataset(Dataset):
    def __init__(self, data, tokenizer, max_length, label_dicts):
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.label_dicts = label_dicts

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data.iloc[idx]
        input_text = item['TITLE_NM'] + " " + item['description']
        encoding = self.tokenizer(
            input_text,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        # KeyError 방지를 위해 `.get()` 사용
        labels = {
            'level1': self.label_dicts['level1'].get(str(item['level1']), 0),
            'level2': self.label_dicts['level2'].get(str(item['level2']), 0),
            'level3': self.label_dicts['level3'].get(str(item['level3']), 0)
        }

        return {
            'input_ids': encoding['input_ids'].squeeze(0),
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'labels': labels
        }

# 모델 정의 (Cascading Classifier)
class CascadingBERTClassifier(nn.Module):
    def __init__(self, bert_model, num_classes):
        super(CascadingBERTClassifier, self).__init__()
        self.bert = bert_model
        self.fc1 = nn.Linear(self.bert.config.hidden_size, num_classes['level1'])
        self.fc2 = nn.Linear(self.bert.config.hidden_size + num_classes['level1'], num_classes['level2'])
        self.fc3 = nn.Linear(self.bert.config.hidden_size + num_classes['level2'], num_classes['level3'])

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]

        level1_logits = self.fc1(cls_output)
        level1_pred = F.one_hot(level1_logits.argmax(dim=-1), num_classes=level1_logits.shape[-1]).float()

        level2_input = torch.cat([cls_output, level1_pred], dim=-1)
        level2_logits = self.fc2(level2_input)
        level2_pred = F.one_hot(level2_logits.argmax(dim=-1), num_classes=level2_logits.shape[-1]).float()

        level3_input = torch.cat([cls_output, level2_pred], dim=-1)
        level3_logits = self.fc3(level3_input)

        return level1_logits, level2_logits, level3_logits


# 하이퍼파라미터
MAX_LENGTH = 202
BATCH_SIZE = 32
LEARNING_RATE = 2e-5
EPOCHS = 5
GRAD_ACCUMULATION_STEPS = 4

# 데이터 준비
final_df['KDC_NM'] = final_df['KDC_NM'].astype(str)
tokenizer = AutoTokenizer.from_pretrained("skt/kobert-base-v1")

# KDC_NM에서 level1, level2, level3 추출
final_df["level1"] = final_df["KDC_NM"].astype(str).apply(lambda x: x[:1] + "00" if len(x) >= 1 else "000")
final_df["level2"] = final_df["KDC_NM"].astype(str).apply(lambda x: x[:2] + "0" if len(x) >= 2 else "000")
final_df["level3"] = final_df["KDC_NM"].astype(str).apply(lambda x: x if len(x) >= 3 else "000")

# label_dicts 생성
label_dicts = {
    "level1": {str(k): v for v, k in enumerate(sorted(final_df["level1"].unique()))},
    "level2": {str(k): v for v, k in enumerate(sorted(final_df["level2"].unique()))},
    "level3": {str(k): v for v, k in enumerate(sorted(final_df["level3"].unique()))}
}
dataset = BookDataset(final_df, tokenizer, MAX_LENGTH, label_dicts)

indices = list(range(len(dataset)))  # dataset의 길이에 해당하는 index 리스트 생성

train_indices, temp_indices = train_test_split(
    indices, test_size=0.2, random_state=42  # stratify 제외
)
val_indices, test_indices = train_test_split(
    temp_indices, test_size=0.5, random_state=42
)

train_dataset = Subset(dataset, train_indices)
val_dataset = Subset(dataset, val_indices)
test_dataset = Subset(dataset, test_indices)

# DataLoader 생성
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, pin_memory=True)

# 데이터셋에서 라벨 추출
# 데이터셋에서 라벨 추출 (수정된 부분)
train_labels = [sample['labels']['level3'] for sample in train_dataset] # 'labels' key and 'level3' subkey used
val_labels = [sample['labels']['level3'] for sample in val_dataset]   # 'labels' key and 'level3' subkey used
test_labels = [sample['labels']['level3'] for sample in test_dataset]  # 'labels' key and 'level3' subkey used

# 라벨 분포 시각화
plt.figure(figsize=(12, 4))

# Train 데이터셋 라벨 분포
plt.subplot(1, 3, 1)
plt.hist(train_labels, bins=np.arange(min(train_labels), max(train_labels)+1, 1), edgecolor='black', alpha=0.7)
plt.title('Train Label Distribution')
plt.xlabel('Label')
plt.ylabel('Frequency')

# Validation 데이터셋 라벨 분포
plt.subplot(1, 3, 2)
plt.hist(val_labels, bins=np.arange(min(val_labels), max(val_labels)+1, 1), edgecolor='black', alpha=0.7)
plt.title('Validation Label Distribution')
plt.xlabel('Label')
plt.ylabel('Frequency')

# Test 데이터셋 라벨 분포
plt.subplot(1, 3, 3)
plt.hist(test_labels, bins=np.arange(min(test_labels), max(test_labels)+1, 1), edgecolor='black', alpha=0.7)
plt.title('Test Label Distribution')
plt.xlabel('Label')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

def compute_loss(outputs, labels, loss_fn):
    label1 = torch.tensor(labels['level1']).to(outputs[0].device)
    label2 = torch.tensor(labels['level2']).to(outputs[1].device)
    label3 = torch.tensor(labels['level3']).to(outputs[2].device)

    loss1 = loss_fn(outputs[0], label1)
    loss2 = loss_fn(outputs[1], label2)
    loss3 = loss_fn(outputs[2], label3)

    return loss1 + loss2 + loss3

# 레이블별 개수 계산
from collections import Counter

labels = dataset.data['level3'].astype(str).tolist()
label_counts = Counter([dataset.label_dicts['level3'].get(label, 0) for label in labels])

# 클래스 개수 수정
num_classes = len(label_dicts['level3'])
total_samples = sum(label_counts.values())

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 가중치 계산 (0.1로 기본값 설정하여 오류 방지)
class_weights = [total_samples / (num_classes * label_counts.get(i, 0.1)) for i in range(num_classes)]
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(device)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm  # Jupyter 환경 등에서 적절한 tqdm 사용
from sklearn.metrics import f1_score, roc_auc_score, recall_score, accuracy_score, precision_score

# 예시: 다중 분류 각 레벨별 loss를 합산하는 함수 (모델 구조에 맞게 수정 필요)
def compute_loss(outputs, labels, loss_fn):
    total_loss = 0
    for i, level in enumerate(["level1", "level2", "level3"]):
        total_loss += loss_fn(outputs[i], labels[level])
    return total_loss

# 학습 함수
def train_model(model, train_loader, val_loader, optimizer, loss_fn, device, num_epochs, patience):
    model.to(device)

    train_losses, val_losses = [], []
    train_accuracies, val_accuracies = [], []
    best_val_loss = float("inf")
    early_stopping_counter = 0

    for epoch in range(num_epochs):
        # Train Mode
        model.train()
        total_loss = 0
        corrects = {"level1": 0, "level2": 0, "level3": 0}
        total = 0

        # leave=True: 에포크 완료 후 진행바를 화면에 남기고, leave=False로 하면 한 에포크당 진행바가 하나만 보입니다.
        train_tqdm = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Training", leave=False)

        for batch in train_tqdm:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = {key: batch['labels'][key].to(device) for key in batch['labels']}

            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask)
            loss = compute_loss(outputs, labels, loss_fn)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            # 각 레벨별 예측 결과 계산 (모델 출력 구조에 맞게 조정)
            for level in corrects.keys():
                preds = outputs[int(level[-1]) - 1].argmax(dim=-1)
                corrects[level] += (preds == labels[level]).sum().item()
            total += len(batch['input_ids'])

            train_tqdm.set_postfix(loss=loss.item())

        train_loss = total_loss / len(train_loader)
        train_losses.append(train_loss)
        train_accuracy = sum(corrects.values()) / (3 * total)  # 3계층 평균 정확도
        train_accuracies.append(train_accuracy)

        # Validation Mode
        model.eval()
        val_loss, val_accuracy = evaluate_model(model, val_loader, loss_fn, device)
        val_losses.append(val_loss)
        val_accuracies.append(val_accuracy)

        print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Train Acc: {train_accuracy:.4f}")
        print(f"Epoch {epoch+1}/{num_epochs} - Val Loss: {val_loss:.4f}, Val Acc: {val_accuracy:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), "best_model.pt")
            early_stopping_counter = 0
        else:
            early_stopping_counter += 1

        if early_stopping_counter >= patience:
            print("Early stopping triggered!")
            break

    # 학습 과정 그래프 시각화
    plt.figure(figsize=(12, 5))

    # Loss 그래프
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(train_losses)+1), train_losses, label="Train Loss", marker="o")
    plt.plot(range(1, len(val_losses)+1), val_losses, label="Validation Loss", marker="o")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Train vs Validation Loss")
    plt.legend()

    # Accuracy 그래프
    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(train_accuracies)+1), train_accuracies, label="Train Accuracy", marker="o")
    plt.plot(range(1, len(val_accuracies)+1), val_accuracies, label="Validation Accuracy", marker="o")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Train vs Validation Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.show()

# 평가 함수
def evaluate_model(model, dataloader, loss_fn, device):
    model.eval()
    total_loss = 0
    corrects = {"level1": 0, "level2": 0, "level3": 0}
    total = 0

    val_tqdm = tqdm(dataloader, desc="Validating", leave=False)
    with torch.no_grad():
        for batch in val_tqdm:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = {key: batch['labels'][key].to(device) for key in batch['labels']}

            outputs = model(input_ids, attention_mask)
            loss = compute_loss(outputs, labels, loss_fn)
            total_loss += loss.item()

            for level in corrects.keys():
                preds = outputs[int(level[-1]) - 1].argmax(dim=-1)
                corrects[level] += (preds == labels[level]).sum().item()
            total += len(batch['input_ids'])
    avg_loss = total_loss / len(dataloader)
    accuracy = sum(corrects.values()) / (3 * total)
    return avg_loss, accuracy

# 테스트 평가 함수
def test_model(model, test_loader, device):
    model.eval()
    all_preds = {"level1": [], "level2": [], "level3": []}
    all_labels = {"level1": [], "level2": [], "level3": []}

    test_tqdm = tqdm(test_loader, desc="Testing", leave=False)
    with torch.no_grad():
        for batch in test_tqdm:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = {key: batch['labels'][key].cpu().numpy() for key in batch['labels']}

            outputs = model(input_ids, attention_mask)
            preds = {key: outputs[int(key[-1]) - 1].argmax(dim=-1).cpu().numpy() for key in labels.keys()}

            for key in labels.keys():
                all_labels[key].extend(labels[key])
                all_preds[key].extend(preds[key])

    # 성능 평가
    metrics = {}
    for key in all_preds.keys():
        metrics[key] = {
            "Accuracy": accuracy_score(all_labels[key], all_preds[key]),
            "Micro F1": f1_score(all_labels[key], all_preds[key], average="micro"),
            "Macro F1": f1_score(all_labels[key], all_preds[key], average="macro"),
            "Micro Precision": precision_score(all_labels[key], all_preds[key], average="micro"),
            "Macro Precision": precision_score(all_labels[key], all_preds[key], average="macro"),
            "Micro Recall": recall_score(all_labels[key], all_preds[key], average="micro"),
            "Macro Recall": recall_score(all_labels[key], all_preds[key], average="macro"),
            "Micro ROC AUC": roc_auc_score(all_labels[key], all_preds[key], multi_class="ovo", average="micro"),
            "Macro ROC AUC": roc_auc_score(all_labels[key], all_preds[key], multi_class="ovo", average="macro"),
        }

    print("\nTest Performance:")
    for key, values in metrics.items():
        print(f" - {key}:")
        for metric, value in values.items():
            print(f"   {metric}: {value:.4f}")

    return metrics

# 하이퍼파라미터 및 기타 설정
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
NUM_EPOCHS = 5
LEARNING_RATE = 2e-5
PATIENCE = 3  # early stopping 기준

# 모델 및 옵티마이저 설정
# (BERT 모델과 분류기는 환경에 맞게 미리 정의되어 있어야 합니다)
from transformers import BertModel
# from your_model_module import CascadingBERTClassifier  # 사용자 정의 모델
bert_model = AutoModel.from_pretrained("skt/kobert-base-v1")

# 예시: 데이터프레임 final_df에서 각 레벨의 클래스 수를 구함 (정의되어 있어야 함)
num_classes = {
    "level1": final_df["level1"].nunique(),
    "level2": final_df["level2"].nunique(),
    "level3": final_df["level3"].nunique()
}
model = CascadingBERTClassifier(bert_model, num_classes).to(DEVICE)
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE)
loss_fn = nn.CrossEntropyLoss()

# 모델 학습
train_model(model, train_loader, val_loader, optimizer, loss_fn, DEVICE, NUM_EPOCHS, PATIENCE)

# 테스트 평가
test_metrics = test_model(model, test_loader, DEVICE)
```
